```python
#Modelagem Dimensional - Modelo Estrela
#Este notebook tem como objetivo transformar os dados tratados em um modelo dimensional (Star Schema), pronto para análise em ferramentas de BI como Power BI.
```

```python
import pandas as pd
```

```python
df = pd.read_csv('../data/processed/dados_tratados.csv')

df.head()
```

```python
df.columns
```

```python
dim_paciente = df[['id_paciente', 'idade', 'sexo', 'faixa_etaria']] \
    .drop_duplicates() \
    .reset_index(drop=True)

dim_paciente.head()
```

```python
dim_paciente = df[['id_paciente', 'idade', 'sexo', 'faixa_etaria']] \
    .drop_duplicates() \
    .reset_index(drop=True)

dim_paciente.head()
```

```python
dim_tempo = df[['data_entrada', 'ano', 'mes', 'dia']] \
    .drop_duplicates() \
    .sort_values('data_entrada') \
    .reset_index(drop=True)

dim_tempo.columns = ['data', 'ano', 'mes', 'dia']

dim_tempo.head()
```

```python
dim_tempo = df[['data_entrada', 'ano', 'mes', 'dia']] \
    .drop_duplicates() \
    .sort_values('data_entrada') \
    .reset_index(drop=True)

dim_tempo.columns = ['data', 'ano', 'mes', 'dia']

dim_tempo.head()
```

```python
dim_procedimento = df[['id_procedimento', 'descricao']] \
    .drop_duplicates() \
    .reset_index(drop=True)

dim_procedimento.columns = ['id_procedimento', 'descricao_procedimento']
```

```python
df.columns
```

```python
dim_setor = df[['setor']] \
    .drop_duplicates() \
    .reset_index(drop=True)

dim_setor['id_setor'] = dim_setor.index + 1

dim_setor.head()
```

```python
df = df.merge(dim_setor, on='setor', how='left')
```

```python
fato = df[['id_internacao',
           'id_paciente',
           'id_procedimento',
           'id_setor',
           'data_entrada',
           'custo_total',
           'quantidade',
           'tempo_internacao']]

fato.head()
```

```python
dim_paciente.to_csv('../data/processed/dim_paciente.csv', index=False)
dim_tempo.to_csv('../data/processed/dim_tempo.csv', index=False)
dim_setor.to_csv('../data/processed/dim_setor.csv', index=False)
dim_procedimento.to_csv('../data/processed/dim_procedimento.csv', index=False)
fato.to_csv('../data/processed/fato_internacoes.csv', index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

tabela_resumo = pd.read_csv('../data/processed/tabela_resumo.csv')

plt.figure(figsize=(10,6))

bars = plt.bar(tabela_resumo['Setor'], tabela_resumo['Custo Total'])

# título e labels
plt.title('Custo Total por Setor', fontsize=16, fontweight='bold')
plt.xlabel('Setor')
plt.ylabel('Custo Total')

# rotacionar nomes
plt.xticks(rotation=45)

# adicionar valores nas barras
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, 
             f'{yval:,.0f}', 
             ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

```python
tabela_resumo = tabela_resumo.sort_values(by='Custo Total', ascending=False)
```

```python
tabela_pivot = pd.read_csv('../data/processed/tabela_pivot.csv', index_col=0)

plt.figure(figsize=(10,6))

for setor in tabela_pivot.index:
    plt.plot(tabela_pivot.columns, tabela_pivot.loc[setor], marker='o', label=setor)

plt.title('Evolução de Custos por Setor', fontsize=16, fontweight='bold')
plt.xlabel('Mês')
plt.ylabel('Custo')

plt.legend()
plt.tight_layout()
plt.show()
```

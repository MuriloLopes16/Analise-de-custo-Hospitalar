```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import random
```

```python
# quantidade de pacientes
n_pacientes = 100

pacientes = pd.DataFrame({
    'id_paciente': range(1, n_pacientes + 1),
    'nome': [f'Paciente_{i}' for i in range(1, n_pacientes + 1)],
    'idade': np.random.randint(0, 90, n_pacientes),
    'sexo': np.random.choice(['M', 'F'], n_pacientes)
})

pacientes.head()
```

```python
n_internacoes = 200

setores = ['UTI', 'Emergência', 'Enfermaria', 'Centro Cirúrgico']

internacoes = []

for i in range(1, n_internacoes + 1):
    paciente_id = random.randint(1, n_pacientes)
    
    data_entrada = datetime(2024, 1, 1) + timedelta(days=random.randint(0, 365))
    tempo_internado = random.randint(1, 15)
    data_saida = data_entrada + timedelta(days=tempo_internado)

    setor = random.choice(setores)

    internacoes.append([
        i, paciente_id, data_entrada, data_saida, setor
    ])

internacoes = pd.DataFrame(internacoes, columns=[
    'id_internacao', 'id_paciente', 'data_entrada', 'data_saida', 'setor'
])

internacoes.head()
```

```python
procedimentos = pd.DataFrame({
    'id_procedimento': [1,2,3,4,5],
    'descricao': [
        'Consulta médica',
        'Exame de sangue',
        'Raio-X',
        'Cirurgia',
        'Diária hospitalar'
    ],
    'custo_unitario': [150, 80, 200, 5000, 300]
})

procedimentos
```

```python
faturamento = []

for _, row in internacoes.iterrows():
    qtd_procedimentos = random.randint(1, 5)
    
    for _ in range(qtd_procedimentos):
        proc = procedimentos.sample(1).iloc[0]
        
        quantidade = random.randint(1, 3)
        custo_total = quantidade * proc['custo_unitario']
        
        faturamento.append([
            row['id_internacao'],
            proc['id_procedimento'],
            quantidade,
            custo_total
        ])

faturamento = pd.DataFrame(faturamento, columns=[
    'id_internacao', 'id_procedimento', 'quantidade', 'custo_total'
])

faturamento.head()
```

```python
pacientes.to_csv('data/raw/pacientes.csv', index=False)
internacoes.to_csv('data/raw/internacoes.csv', index=False)
procedimentos.to_csv('data/raw/procedimentos.csv', index=False)
faturamento.to_csv('data/raw/faturamento.csv', index=False)
```
